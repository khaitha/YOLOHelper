```python
import cv2
import os
import matplotlib.pyplot as plt
import random
#--------- Code Written by Khai Tha ----------------
# Folders containing the images and labels
image_folder = 'Final_Dataset/train/images'
label_folder = 'Final_Dataset/train/labels'

# Function to draw bounding boxes on an image
def draw_bounding_boxes(image_path, label_path):
    # Read the image
    image = cv2.imread(image_path)
    h, w, _ = image.shape

    # Read the label file
    with open(label_path, 'r') as file:
        lines = file.readlines()
    
    # Iterate through each bounding box
    total_line = 0
    for line in lines:
        data = line.split()
        class_id = int(data[0])
        x_center, y_center, width, height = map(float, data[1:])

        # Convert normalized coordinates to pixel values
        x_center = int(x_center * w)
        y_center = int(y_center * h)
        width = int(width * w)
        height = int(height * h)

        # Calculate the top-left corner of the bounding box
        x_min = int(x_center - width / 2)
        y_min = int(y_center - height / 2)

        # Draw the rectangle on the image
        cv2.rectangle(image, (x_min, y_min), (x_min + width, y_min + height), (255, 0, 0), 1)
        total_line += 1
    print(total_line)
    return image

# Get all image filenames
image_filenames = [f for f in os.listdir(image_folder) if f.endswith('.jpg') or f.endswith('.png')]
randomnumber1 = random.randint(0,1598)
randomnumber2 = randomnumber1 + 30
# Select the first 5 filenames
first_five_filenames = image_filenames[randomnumber1:randomnumber2]

# Visualize the bounding boxes for the selected images
for filename in first_five_filenames:
    image_path = os.path.join(image_folder, filename)
    label_path = os.path.join(label_folder, os.path.splitext(filename)[0] + '.txt')

    if os.path.exists(label_path):
        print(f"Image: {filename}, Label: {os.path.splitext(filename)[0]}.txt")
        image_with_boxes = draw_bounding_boxes(image_path, label_path)

        # Resize the image to be smaller
        #image_with_boxes = cv2.resize(image_with_boxes, (0, 0), fx=0.5, fy=0.5)

        # Convert BGR to RGB for matplotlib
        image_rgb = cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB)

        # Display the image
        plt.figure(figsize=(20, 20))
        plt.imshow(image_rgb)
        plt.axis('off')
        plt.show()
```
